```python
#To find out which store has maximum sales
sales_list=[]
sales_list=df.groupby(['Store'])['Weekly_Sales'].sum()
max_sales=max(df.groupby(['Store'])['Weekly_Sales'].sum())
sales_list
```

```python
for i in range(1,46):
    if max_sales==sales_list[i]:
        print("Store which has maximum sales of {} is {}".format(max_sales,i))
```

```python
#from the above result we see the STORE 20 has maximum sales
```

```python
#to find the standard devitation
std_dev=[]
std_dev=df.groupby(['Store'])['Weekly_Sales'].std()
max_std=max(df.groupby(['Store'])['Weekly_Sales'].std())
print(std_dev)
```

```python
for i in range(1,46):
    if max_std==std_dev[i]:
        print('Store which has maximum standard deviation of {} is {}'.format(max_std,i))
```

```python
#from the above we see the STORE 14 has maximum std
```

```python
#Coefficient of mean to std
df_var_sales = df.groupby('Store')['Weekly_Sales'].var().sort_values(ascending=False)
df_var_sales.head()
```

```python
#Linear Regression Model
x=df.drop(["Weekly_Sales","Date"],axis=1)
y=df["Weekly_Sales"]
```

```python
linreg=LinearRegression(n_jobs=-1)
xtrain,xtest,ytrain,ytest=model_selection.train_test_split(x,y,test_size=0.4,random_state=42)
linreg.fit(xtrain,ytrain)
```

```python
print(linreg.intercept_)
print(linreg.coef_)
```

```python
x.columns
```

```python
features=['Store', 'Holiday_Flag', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment', 'Year', 'Month']
```

```python
relation=pd.Series(linreg.coef_,x.columns).sort_values()
relation.plot(kind="bar")
```

```python
#The plot shows that fuel price have greater positive impact on weekly sales. 
#Unemployment also has certain negative impact on weekly sales. 
#CPI has least impact towards weekly sales.
```

```python
print(format(linreg.score(xtest,ytest)))
```

```python
print(sqrt(mean_squared_error(ytrain,linreg.predict(xtrain))))
```

```python
print(sqrt(mean_squared_error(ytest,linreg.predict(xtest))))
```

```python
#From the above observations: Thus test error is less compared to train error. Hence our predicted model is well and good
```
